```python
# =============================================================================
# TADP: Trustworthy AI Data Preparation Governance & Evaluation Pipeline
# =============================================================================

import os, io, glob, time, uuid, shutil, json, itertools
import warnings
from datetime import datetime, timezone
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Quiet noisy libs
warnings.filterwarnings('ignore', category=DeprecationWarning)
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings("ignore", category=DeprecationWarning, module="jupyter_client.session")

# Suppress TensorFlow GPU warnings and force CPU
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
os.environ['CUDA_VISIBLE_DEVICES'] = '-1'

# Import ML libraries
try:
    import tensorflow as tf
    TF_AVAILABLE = True
except ImportError:
    print("âš ï¸ TensorFlow not available, installing...")
    import subprocess
    subprocess.run(["pip", "install", "tensorflow"], check=True)
    import tensorflow as tf
    TF_AVAILABLE = True

try:
    import flwr as fl
    FLWR_AVAILABLE = True
    print("âœ… Flower successfully imported")
except ImportError as e:
    FLWR_AVAILABLE = False
    print(f"âš ï¸ Flower import failed: {e}")

try:
    from IPython.display import display, HTML
    IPYTHON_AVAILABLE = True
except ImportError:
    IPYTHON_AVAILABLE = False
    def display(*args, **kwargs):
        for arg in args:
            print(arg)
    def HTML(text):
        return text

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, MaxAbsScaler
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# =========================
# Global Configuration
# =========================
NUM_CLIENTS = 3
RANDOM_STATE = 42
NUM_ROUNDS_FL = 5
DEFAULT_LOCAL_EPOCHS = 1
DEFAULT_CLIENT_FRACTION = 0.5  # C = 0.5 for partial participation

# Energy and cost parameters
POWER_W = 45.0
COST_PER_KWH = 0.2
CARBON_INTENSITY = 0.475

# Protocol overhead
CENTRAL_PROTOCOL_OVERHEAD = 0.10
FED_PROTOCOL_OVERHEAD = 0.10

# Governance cutoffs
HIGH_CUT = 3.0
GOOD_CUT = 2.25

# P-score weights
WEIGHTS_PSCORE = {'dim1': 0.25, 'dim2': 0.15, 'dim3': 0.10,
                  'dim4': 0.10, 'dim5': 0.30, 'dim6': 0.10}

# File paths
CENTRAL_LEDGER_PATH = "./ledgers/central/central_ledger.csv"
OVERHEAD_PATH = "./ledgers/overhead.csv"
LEDGER_GROWTH_PATH = "./ledgers/storage_metrics.csv"
STORAGE_TRACKING_PATH = "./ledgers/storage_tracking.csv"

np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)

# Enhanced Configuration
FEDPROX_MU = 0.01  # Proximal term coefficient for FedProx

# =========================
# Core Data Functions
# =========================
def load_diabetes_csv():
    """Load diabetes dataset from file or upload"""
    if os.path.exists("diabetes.csv"):
        return pd.read_csv("diabetes.csv")
    try:
        from google.colab import files
        print("â¬†ï¸ Upload diabetes.csv")
        uploaded = files.upload()
        fname = next(iter(uploaded))
        return pd.read_csv(io.BytesIO(uploaded[fname]))
    except Exception:
        raise FileNotFoundError("Place diabetes.csv next to this script or upload it in Colab.")

def preprocess_diabetes(df):
    """Preprocess diabetes dataset"""
    df = df.copy()
    df.columns = df.columns.str.lower().str.replace('[^a-z0-9]+','_', regex=True)
    df.replace('?', np.nan, inplace=True)

    # Preprocess age
    if 'age' in df and df['age'].dtype == object:
        df['age'] = df['age'].str.extract(r'(\d+)').astype(float)

    # Preprocess target
    if 'readmitted' in df:
        df['readmitted'] = df['readmitted'].map({'NO': 0, '<30': 1, '>30': 2})
        df['readmitted'] = df['readmitted'].fillna(0).astype(int)

    # Clean diagnosis codes
    for col in ['diag_1', 'diag_2', 'diag_3']:
        if col in df:
            df[col] = df[col].replace({'V[0-9]{2}': np.nan, 'E[0-9]{3}': np.nan}, regex=True)
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

    # Handle missing values
    for col in df.columns:
        if df[col].dtype in ['float64', 'int64']:
            df[col] = df[col].fillna(0)
        elif df[col].dtype == 'object':
            df[col] = df[col].fillna('Unknown')

    return df

def split_clients(df, num_clients=NUM_CLIENTS):
    """Split data into non-IID client partitions"""
    data_main = df.sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)
    fracs = list(np.random.dirichlet([1.5]*num_clients))
    client_ids = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")[:num_clients]
    client_data = {}
    start = 0

    for cid in client_ids:
        if not fracs: break
        frac = fracs.pop()
        end = start + int(frac*len(data_main))
        client_data[cid] = data_main.iloc[start:end].copy()
        start = end

    if start < len(data_main):
        client_data[client_ids[-1]] = pd.concat([client_data[client_ids[-1]], data_main.iloc[start:]])

    return client_ids, client_data

def preprocess_data(client_data, target_col="readmitted"):
    """Preprocess data for machine learning"""
    df_all = pd.concat(client_data.values()).reset_index(drop=True)
    X = df_all.drop(target_col, axis=1)
    y = df_all[target_col]

    X_tr_raw, X_te_raw, y_tr, y_te = train_test_split(
        X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
    )

    num_cols = X.select_dtypes(include=np.number).columns.tolist()
    cat_cols = X.select_dtypes(exclude=np.number).columns.tolist()

    try:
        ohe = OneHotEncoder(sparse_output=True, handle_unknown='ignore')
    except TypeError:
        ohe = OneHotEncoder(sparse=True, handle_unknown='ignore')

    preprocessor = ColumnTransformer([
        ('num', MaxAbsScaler(), num_cols),
        ('cat', ohe, cat_cols)
    ], sparse_threshold=1.0)

    X_train = preprocessor.fit_transform(X_tr_raw)
    X_test = preprocessor.transform(X_te_raw)

    if hasattr(X_train, "toarray"):
        X_train = X_train.toarray()
        X_test = X_test.toarray()

    if np.isnan(X_train).any():
        X_train = np.nan_to_num(X_train)
        X_test = np.nan_to_num(X_test)

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    return X_train, X_test, y_tr, y_te, preprocessor, cv

# =========================
# Filesystem and Ledger Management
# =========================
def ensure_dirs():
    """Ensure required directories exist"""
    os.makedirs("./ledgers", exist_ok=True)
    os.makedirs("./ledgers/central", exist_ok=True)
    os.makedirs("./ledgers/local", exist_ok=True)
    os.makedirs("./ledgers/scenarios", exist_ok=True)

def reset_ledgers():
    """Reset all ledger files"""
    if os.path.exists("./ledgers"):
        shutil.rmtree("./ledgers")
    ensure_dirs()

    # Central ledger
    pd.DataFrame(columns=[
        "tx_id","timestamp","scenario","approach","mode","job_idx",
        "client","version","record_count","dim1","dim2","dim3","dim4","dim5","dim6",
        "pscore","initial_action","action"
    ]).to_csv(CENTRAL_LEDGER_PATH, index=False)

    # Overhead ledger
    pd.DataFrame(columns=[
        "tx_id","timestamp","scenario","approach","mode","job_idx","round","client",
        "params","upload_MB","download_MB","fit_time_s","overhead_write_time_s"
    ]).to_csv(OVERHEAD_PATH, index=False)

    # Storage metrics
    pd.DataFrame(columns=[
        "timestamp","stage","scenario","round","client","central_MB","local_total_MB","overhead_MB","total_MB"
    ]).to_csv(LEDGER_GROWTH_PATH, index=False)

    # Storage tracking
    pd.DataFrame(columns=[
        "timestamp", "scenario", "round", "client", "operation",
        "central_size_mb", "local_size_mb", "overhead_size_mb", "total_size_mb"
    ]).to_csv(STORAGE_TRACKING_PATH, index=False)

    print("âœ… Global ledgers reset.")

def _sizeof_mb(path: str) -> float:
    """Get file size in MB"""
    try:
        return os.path.getsize(path)/(1024**2)
    except Exception:
        return 0.0

def track_storage_growth(scenario, round_num, client, operation):
    """Track storage growth for each operation"""
    timestamp = datetime.now(timezone.utc).isoformat()

    central_size = _sizeof_mb(CENTRAL_LEDGER_PATH)
    local_size = sum(_sizeof_mb(f"./ledgers/local/{c}_ledger.csv") for c in client if os.path.exists(f"./ledgers/local/{c}_ledger.csv"))
    overhead_size = _sizeof_mb(OVERHEAD_PATH)
    total_size = central_size + local_size + overhead_size

    # Log to storage tracking
    row = [timestamp, scenario, round_num, client, operation,
           central_size, local_size, overhead_size, total_size]

    columns = ["timestamp", "scenario", "round", "client", "operation",
               "central_size_mb", "local_size_mb", "overhead_size_mb", "total_size_mb"]

    pd.DataFrame([row], columns=columns).to_csv(
        STORAGE_TRACKING_PATH, mode='a',
        header=not os.path.exists(STORAGE_TRACKING_PATH), index=False
    )

    return central_size, local_size, overhead_size, total_size

# =========================
# Model Utilities
# =========================
def make_model(input_dim, verbose=False):
    """Create and compile neural network model"""
    model = tf.keras.Sequential([
        tf.keras.layers.InputLayer(shape=(input_dim,)),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(3, activation='softmax'),
    ])
    model.compile(
        optimizer=tf.keras.optimizers.Adam(epsilon=1e-7),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    if verbose:
        model.summary()
    return model

def get_detailed_metrics(y_true, y_proba):
    """Calculate comprehensive evaluation metrics"""
    if len(y_true) == 0:
        return {'accuracy': 0, 'precision': 0, 'recall': 0, 'f1': 0, 'roc_auc': 0}, np.zeros((3,3), int)

    y_pred = np.argmax(y_proba, axis=1)
    cm = confusion_matrix(y_true, y_pred, labels=[0,1,2])

    metrics = {
        'accuracy': accuracy_score(y_true, y_pred),
        'precision': precision_score(y_true, y_pred, average='weighted', zero_division=0),
        'recall': recall_score(y_true, y_pred, average='weighted', zero_division=0),
        'f1': f1_score(y_true, y_pred, average='weighted', zero_division=0),
        'roc_auc': roc_auc_score(y_true, y_proba, multi_class='ovr') if len(np.unique(y_true)) > 1 else 0
    }
    return metrics, cm

# =========================
# Energy and Cost Calculations
# =========================
def calculate_energy_cost(runtime_seconds, num_clients=1, power_watts=POWER_W,
                         cost_per_kwh=COST_PER_KWH, carbon_intensity=CARBON_INTENSITY):
    """Calculate energy consumption and costs"""
    energy_wh = power_watts * runtime_seconds / 3600
    cost_usd = energy_wh * cost_per_kwh / 1000
    co2_kg = energy_wh * carbon_intensity / 1000
    return {
        'energy_wh': energy_wh,
        'cost_usd': cost_usd,
        'co2_kg': co2_kg,
        'runtime_s': runtime_seconds
    }

def calculate_federated_communication(param_count, num_rounds, effective_clients,
                                    overhead_rate=FED_PROTOCOL_OVERHEAD):
    """Calculate federated learning communication costs"""
    param_bytes = param_count * 4
    per_round_per_client = (param_bytes + param_bytes) * (1.0 + float(overhead_rate))
    total_bytes = per_round_per_client * int(effective_clients) * int(num_rounds)
    return total_bytes / (1024**2)  # MB

def estimate_dataset_size_mb(client_data_dict):
    """Estimate dataset sizes in MB"""
    sizes = {}
    for cid, df in client_data_dict.items():
        total_bytes = len(df) * len(df.columns) * 100  # Approximate bytes per value
        sizes[cid] = total_bytes / (1024**2)
    return sizes

# =========================
# Governance Components
# =========================
quality_profiles = {
    'high': {
        'dim1': {'source': 5.0, 'audits': 5.0, 'collection': 5.0},
        'dim2': {'completeness': 5.0, 'errors': 5.0},
        'dim3': {'icd10_docs': 5.0, 'protocols': 5.0},
        'dim4': {'freshness': 5.0, 'retention': 5.0},
        'dim5': {'hipaa': 5.0, 'ethics': 5.0},
        'dim6': {'license_terms': 5.0, 'redistribution': 5.0, 'ethical_reviews': 5.0}
    },
    'moderate': {
        'dim1': {'source': 3.0, 'audits': 3.0, 'collection': 3.0},
        'dim2': {'completeness': 3.0, 'errors': 3.0},
        'dim3': {'icd10_docs': 3.0, 'protocols': 3.0},
        'dim4': {'freshness': 3.0, 'retention': 3.0},
        'dim5': {'hipaa': 3.0, 'ethics': 3.0},
        'dim6': {'license_terms': 3.0, 'redistribution': 3.0, 'ethical_reviews': 3.0}
    },
    'low': {
        'dim1': {'source': 1.0, 'audits': 1.0, 'collection': 1.0},
        'dim2': {'completeness': 1.0, 'errors': 1.0},
        'dim3': {'icd10_docs': 1.0, 'protocols': 1.0},
        'dim4': {'freshness': 1.0, 'retention': 1.0},
        'dim5': {'hipaa': 1.0, 'ethics': 1.0},
        'dim6': {'license_terms': 1.0, 'redistribution': 1.0, 'ethical_reviews': 1.0}
    }
}

def _mean_or_val(x):
    """Calculate mean of dictionary values or return float value"""
    if isinstance(x, dict):
        vals = [v for v in x.values() if pd.notna(v)]
        return float(np.mean(vals)) if vals else 0.0
    try:
        return float(x)
    except:
        return 0.0

def assign_quality_profile(client_id, client_data):
    """Assign quality profile to client"""
    profiles = list(quality_profiles.keys())
    weights = [0.2, 0.5, 0.3]
    profile = np.random.choice(profiles, p=weights)
    return profile, quality_profiles[profile]

def calculate_medical_quality(df):
    """Calculate medical data quality metrics"""
    qm = {'completeness': 0, 'errors': 0}
    crit_fields = ['age', 'diag_1', 'max_glu_serum', 'a1cresult', 'readmitted']
    crit = [c for c in crit_fields if c in df]

    if crit:
        qm['completeness'] = (1 - df[crit].isna().mean().mean()) * 5

    vc = []
    if 'diag_1' in df:
        vc.append(0.6 * df['diag_1'].astype(str).str.match(r'\d{3}').mean())
    if 'time_in_hospital' in df:
        vc.append(0.2 * (df['time_in_hospital'] >= 1).mean())
    if 'num_medications' in df:
        vc.append(0.2 * (df['num_medications'] > 0).mean())

    if vc:
        qm['errors'] = max(0.0, (1 - sum(vc))) * 5

    return qm

def calculate_usage_constraints(df):
    """Calculate usage constraint metrics"""
    n = len(df)
    defaults = pd.Series([3] * n)
    t = df.get("license_terms", defaults).dropna().astype(float).mean()
    r = df.get("redistribution", defaults).dropna().astype(float).mean()
    e = df.get("ethical_reviews", defaults).dropna().astype(float).mean()
    return {"license_terms": t, "redistribution": r, "ethical_reviews": e}

def calculate_pscore(client_id, df):
    """Calculate comprehensive P-score for client data"""
    profile, dim_scores = assign_quality_profile(client_id, df)

    # Calculate dimension values
    dim1 = _mean_or_val(dim_scores.get('dim1', 0.0))
    dim2 = _mean_or_val(calculate_medical_quality(df))
    dim3 = _mean_or_val(dim_scores.get('dim3', 0.0))
    dim4 = _mean_or_val(dim_scores.get('dim4', 0.0))
    dim5 = _mean_or_val(dim_scores.get('dim5', 0.0))
    dim6 = _mean_or_val(calculate_usage_constraints(df))

    dim_vals = [dim1, dim2, dim3, dim4, dim5, dim6]
    wts = [WEIGHTS_PSCORE[f'dim{i+1}'] for i in range(6)]
    pscore = sum(d * v for d, v in zip(dim_vals, wts))

    return pscore, dim_vals, profile

# =========================
# Governance Decision Engine
# =========================
def dynamic_cutoffs(pscores, template):
    """Calculate dynamic cutoffs based on template"""
    if template == 'AA':
        return (0.0, 0.0)  # Accept All
    elif template == 'VR':  # Value Ranking
        s = sorted(pscores, reverse=True)
        high_cut = s[int(len(s)*0.2)-1] if len(s) >= 5 else s[0] if s else 5.0
        good_cut = s[int(len(s)*0.8)-1] if len(s) >= 5 else s[-1] if s else 0.0
        return (high_cut, good_cut)
    elif template == 'SDA':  # Static-Dynamic-Adjust
        s = sorted(pscores, reverse=True)
        if len(s) < 2:
            return (5.0, 0.0)
        high_cut = (s[0] + s[1]) / 2.0
        good_cut = s[1] + 1e-5
        return (high_cut, good_cut)
    else:
        return (HIGH_CUT, GOOD_CUT)

def decide_action(pscore, high_cut, good_cut, template):
    """Decide governance action based on P-score"""
    if template == 'AA':
        return 'ACCEPT'
    if pscore >= high_cut:
        return 'ACCEPT'
    if pscore >= good_cut:
        return 'REVIEW'
    return 'QUARANTINE'

def manual_review(client_id, pscore, high_cut, good_cut):
    """Manual review for datasets in REVIEW state"""
    print(f"\nğŸ” MANUAL REVIEW REQUIRED for Client {client_id}")
    print(f"   P-score: {pscore:.2f} (High Cut: {high_cut:.2f}, Good Cut: {good_cut:.2f})")
    print(f"   This dataset requires manual review decision.")

    while True:
        decision = input("   Enter 'a' to ACCEPT or 'r' to REJECT: ").strip().lower()
        if decision in ['a', 'accept']:
            return 'ACCEPT'
        elif decision in ['r', 'reject']:
            return 'QUARANTINE'
        else:
            print("   Invalid input. Please enter 'a' for ACCEPT or 'r' for REJECT.")

def simulate_human_review(initial_action, client_id, df, pscore, high_cut, good_cut, manual_mode=False):
    """Simulate human review process - now with manual mode"""
    if initial_action != 'REVIEW':
        return initial_action

    if manual_mode:
        return manual_review(client_id, pscore, high_cut, good_cut)
    else:
        # Auto-review for non-manual mode
        if np.random.random() < 0.7:
            return 'ACCEPT'
        return 'QUARANTINE'

# =========================
# Ledger Logging
# =========================
version_counters = {}

def get_next_version(client):
    """Get next version number for client"""
    version_counters[client] = version_counters.get(client, 0) + 1
    return version_counters[client]

def log_to_ledger(client, df, pscore, dim_vals, initial_action, action, scenario, approach, mode, job_idx):
    """Log transaction to central and local ledgers"""
    ensure_dirs()
    tx_id = str(uuid.uuid4())
    timestamp = datetime.now(timezone.utc).isoformat()
    version = get_next_version(client)
    record_count = len(df)
    dim1, dim2, dim3, dim4, dim5, dim6 = dim_vals

    row = [tx_id, timestamp, scenario, approach, mode, job_idx, client, version, record_count,
           dim1, dim2, dim3, dim4, dim5, dim6, pscore, initial_action, action]

    columns = [
        "tx_id", "timestamp", "scenario", "approach", "mode", "job_idx", "client", "version",
        "record_count", "dim1", "dim2", "dim3", "dim4", "dim5", "dim6", "pscore",
        "initial_action", "action"
    ]

    # Log to central ledger
    pd.DataFrame([row], columns=columns).to_csv(
        CENTRAL_LEDGER_PATH, mode='a', header=not os.path.exists(CENTRAL_LEDGER_PATH), index=False
    )

    # Log to local ledger
    local_path = f"./ledgers/local/{client}_ledger.csv"
    pd.DataFrame([row], columns=columns).to_csv(
        local_path, mode='a', header=not os.path.exists(local_path), index=False
    )

    # Track storage growth
    track_storage_growth(scenario, "governance", client, "ledger_update")

    return tx_id

def log_overhead(tx_id, scenario, approach, mode, job_idx, round_num, client, params,
                upload_MB, download_MB, fit_time_s, overhead_write_time_s):
    """Log overhead metrics"""
    ensure_dirs()
    timestamp = datetime.now(timezone.utc).isoformat()
    row = [tx_id, timestamp, scenario, approach, mode, job_idx, round_num, client,
           params, upload_MB, download_MB, fit_time_s, overhead_write_time_s]

    columns = [
        "tx_id", "timestamp", "scenario", "approach", "mode", "job_idx", "round", "client",
        "params", "upload_MB", "download_MB", "fit_time_s", "overhead_write_time_s"
    ]

    pd.DataFrame([row], columns=columns).to_csv(
        OVERHEAD_PATH, mode='a', header=not os.path.exists(OVERHEAD_PATH), index=False
    )

    # Track storage growth
    track_storage_growth(scenario, round_num, client, "overhead_log")

# =========================
# Governance Orchestration
# =========================
def run_tadp_governance(client_ids, client_data, scenario, approach, mode, job_idx, template='VR', manual_review_mode=False):
    """Run TADP governance process"""
    print(f"  Running TADP governance ({template})...")
    pscores = {}
    dims = {}
    profiles = {}

    for cid in client_ids:
        pscore, dim_vals, profile = calculate_pscore(cid, client_data[cid])
        pscores[cid] = pscore
        dims[cid] = dim_vals
        profiles[cid] = profile

    high_cut, good_cut = dynamic_cutoffs(list(pscores.values()), template)
    accepted = []
    decisions = {}

    for cid in client_ids:
        initial_action = decide_action(pscores[cid], high_cut, good_cut, template)
        action = simulate_human_review(initial_action, cid, client_data[cid], pscores[cid], high_cut, good_cut, manual_review_mode)
        decisions[cid] = (initial_action, action, pscores[cid], dims[cid], profiles[cid])

        if action == 'ACCEPT':
            accepted.append(cid)

        log_to_ledger(cid, client_data[cid], pscores[cid], dims[cid], initial_action, action,
                     scenario, approach, mode, job_idx)

    print(f"    Accepted {len(accepted)}/{len(client_ids)} clients")
    return accepted, decisions, (high_cut, good_cut)

def run_baseline_governance(client_ids, client_data, scenario, approach, mode, job_idx):
    """Run baseline (no governance) process"""
    print(f"  Running baseline (no governance)...")
    accepted = client_ids.copy()
    decisions = {}

    for cid in client_ids:
        pscore, dim_vals, profile = calculate_pscore(cid, client_data[cid])
        decisions[cid] = ('ACCEPT', 'ACCEPT', pscore, dim_vals, profile)
        log_to_ledger(cid, client_data[cid], pscore, dim_vals, 'ACCEPT', 'ACCEPT',
                     scenario, approach, mode, job_idx)

    return accepted, decisions, (0.0, 0.0)

# =========================
# Training Components
# =========================
def centralized_training(model, accepted_clients, client_data, scenario, approach, mode, job_idx, client_dataset_sizes):
    """Run centralized training with accepted clients"""
    print(f"  Centralized training with {len(accepted_clients)} clients...")

    # Combine accepted client data
    combined_data = pd.concat([client_data[cid] for cid in accepted_clients])
    X_combined = combined_data.drop("readmitted", axis=1)
    y_combined = combined_data["readmitted"]

    # Transform using global preprocessor
    X_transformed = preprocessor.transform(X_combined)
    if hasattr(X_transformed, "toarray"):
        X_transformed = X_transformed.toarray()

    # FIX: Use same number of epochs as federated for fairness
    # In federated: NUM_ROUNDS_FL * DEFAULT_LOCAL_EPOCHS effective epochs
    # In centralized: Use equivalent total computation
    equivalent_epochs = NUM_ROUNDS_FL * DEFAULT_LOCAL_EPOCHS

    # Train model
    start_time = time.time()
    history = model.fit(
        X_transformed, y_combined.values,
        epochs=equivalent_epochs,  # FIXED: Same total computation as federated
        batch_size=32, verbose=0, validation_split=0.2
    )
    training_time = time.time() - start_time

    # Log overhead
    tx_id = str(uuid.uuid4())
    total_upload_mb = sum(client_dataset_sizes.get(cid, 0) for cid in accepted_clients)
    total_upload_mb *= (1.0 + CENTRAL_PROTOCOL_OVERHEAD)

    log_overhead(tx_id, scenario, approach, mode, job_idx, 0, "central",
                len(model.get_weights()), total_upload_mb, 0, training_time, 0.0)

    return history, model

def federated_training(model, x_test, y_test, accepted_clients, client_data, scenario, approach, mode, job_idx, client_dataset_sizes, k_value=None):
    """Run federated training with accepted clients"""
    print(f"  Federated training with {len(accepted_clients)} clients...")

    global_weights = model.get_weights()
    history_accuracy = []
    cumulative_comm = 0

    for round_num in range(NUM_ROUNDS_FL):
        print(f"    Round {round_num+1}/{NUM_ROUNDS_FL}")
        client_weights = []
        client_sizes = []
        round_comm = 0

        # CLIENT SELECTION LOGIC
        if approach == 'simpleavg':
            # SimpleAvg: Use all clients (no partial participation)
            round_clients = accepted_clients
        elif approach == 'random-k' and k_value is not None:
            # Random-K: Select K random clients each round (reselect every round)
            num_selected = min(k_value, len(accepted_clients))
            round_clients = list(np.random.choice(accepted_clients, size=num_selected, replace=False))
        else:
            # FedAvg/FedProx: Apply client fraction for partial participation
            num_selected = max(1, int(len(accepted_clients) * DEFAULT_CLIENT_FRACTION))
            round_clients = list(np.random.choice(accepted_clients, size=num_selected, replace=False))

        print(f"      Selected {len(round_clients)} clients for this round")

        for cid in round_clients:
            df = client_data[cid]
            X_client = df.drop("readmitted", axis=1)
            y_client = df["readmitted"]

            # Transform using global preprocessor
            X_transformed = preprocessor.transform(X_client)
            if hasattr(X_transformed, "toarray"):
                X_transformed = X_transformed.toarray()

            # Create client model
            client_model = tf.keras.models.clone_model(model)
            client_model.set_weights(global_weights)

            # Train client
            start_time = time.time()
            client_model.fit(X_transformed, y_client.values, epochs=DEFAULT_LOCAL_EPOCHS,
                           batch_size=32, verbose=0)
            fit_time = time.time() - start_time

            # Store weights and dataset size
            client_weights.append(client_model.get_weights())
            client_sizes.append(len(X_transformed))

            # Calculate communication
            params = sum(np.prod(w.shape) for w in client_model.get_weights())
            upload_mb = (params * 4) / (1024**2)
            download_mb = upload_mb
            round_comm += upload_mb + download_mb

            # Log overhead
            tx_id = str(uuid.uuid4())
            log_overhead(tx_id, scenario, approach, mode, job_idx, round_num+1, cid,
                        params, upload_mb, download_mb, fit_time, 0.0)

            # Track storage after each client update
            track_storage_growth(scenario, round_num+1, cid, "client_update")

        cumulative_comm += round_comm

        # FEDERATED AVERAGING - DIFFERENT STRATEGIES
        new_weights = []
        if approach == 'simpleavg':
            # NAÃVE FEDERATED (SimpleAvg): Equal weight averaging (ignore dataset sizes)
            print(f"      Using SimpleAvg: equal weight averaging across {len(client_weights)} clients")
            for i in range(len(global_weights)):
                layer_weights = np.zeros_like(global_weights[i])
                for weights in client_weights:
                    layer_weights += weights[i]
                new_weights.append(layer_weights / len(client_weights))

        elif approach == 'fedprox':
            # FedProx: Sample-size weighted averaging with proximal regularization
            print(f"      Using FedProx: sample-size weighted + proximal term (Î¼={FEDPROX_MU})")
            total_size = sum(client_sizes)

            for i in range(len(global_weights)):
                layer_weights = np.zeros_like(global_weights[i])
                for j, weights in enumerate(client_weights):
                    layer_weights += weights[i] * (client_sizes[j] / total_size)

                # Add proximal regularization: Î¼/2 * ||w - w_global||Â²
                # This encourages client updates to stay close to global model
                proximal_update = global_weights[i] + FEDPROX_MU * (layer_weights - global_weights[i])
                new_weights.append(proximal_update)

        else:
            # VANILLA FedAvg and Random-K: Sample-size weighted averaging
            print(f"      Using FedAvg: sample-size weighted averaging")
            total_size = sum(client_sizes)

            for i in range(len(global_weights)):
                layer_weights = np.zeros_like(global_weights[i])
                for j, weights in enumerate(client_weights):
                    layer_weights += weights[i] * (client_sizes[j] / total_size)
                new_weights.append(layer_weights)

        global_weights = new_weights

        # Evaluate global model
        model.set_weights(global_weights)
        y_proba = model.predict(x_test, verbose=0)
        metrics, _ = get_detailed_metrics(y_test.values, y_proba)
        history_accuracy.append(metrics['accuracy'])

        # Print round summary
        if approach == 'random-k' and k_value is not None:
            print(f"      Round accuracy: {metrics['accuracy']:.3f}, K={len(round_clients)}, Cumulative Comm: {cumulative_comm:.2f} MB")
        else:
            print(f"      Round accuracy: {metrics['accuracy']:.3f}, Clients: {len(round_clients)}/{len(accepted_clients)}, Cumulative Comm: {cumulative_comm:.2f} MB")

    # Set final weights
    model.set_weights(global_weights)

    return model, cumulative_comm

# =========================
# Scenario Execution
# =========================
def run_scenario(name, approach, mode, template, client_ids, client_data, client_dataset_sizes, x_test, y_test, job_idx=1):
    """Execute a single scenario"""
    print(f"\nğŸƒ Running scenario: {name} ({approach}, {mode}, {template})")

    # Run governance
    start_gov = time.time()
    tadp_accepted = None  # Initialize for Random-K case

    if approach in ['baseline', 'simpleavg', 'fedavg', 'fedprox']:
        accepted, decisions, cutoffs = run_baseline_governance(client_ids, client_data, name, approach, mode, job_idx)
    elif approach == 'random-k':
        # Use TADP-VR to determine K (number of clients per round), but RESELECT every round
        tadp_accepted, _, _ = run_tadp_governance(client_ids, client_data, name, 'TADP', mode, job_idx, 'VR')
        k = len(tadp_accepted)  # K = number accepted in TADP-VR
        # For Random-K, we accept all clients but only use K randomly selected each round
        accepted = client_ids.copy()  # All clients are available for selection
        decisions = {}
        for cid in client_ids:
            pscore, dim_vals, profile = calculate_pscore(cid, client_data[cid])
            decisions[cid] = ('ACCEPT', 'ACCEPT', pscore, dim_vals, profile)
        cutoffs = (0.0, 0.0)
        print(f"    Random-K: Will select {k} random clients each round from {len(client_ids)} total")
    else:  # TADP approaches
        # Use manual review only for TADP-VR and TADP-SDA in federated mode
        manual_mode = (template in ['VR', 'SDA']) and (mode == 'federated')
        accepted, decisions, cutoffs = run_tadp_governance(client_ids, client_data, name, approach, mode, job_idx, template, manual_mode)

    gov_time = time.time() - start_gov

    # Initialize model
    input_dim = x_test.shape[1]
    model = make_model(input_dim)

    # Run training
    start_train = time.time()

    if mode == 'centralized':
        # FIX: Use same number of epochs as federated for fairness
        history, trained_model = centralized_training(model, accepted, client_data, name, approach, mode, job_idx, client_dataset_sizes)
        training_time = time.time() - start_train

        # Evaluate
        y_proba = trained_model.predict(x_test, verbose=0)
        metrics, cm = get_detailed_metrics(y_test.values, y_proba)

        # Calculate communication costs
        comm_cost = sum(client_dataset_sizes.get(cid, 0) for cid in accepted)
        comm_cost *= (1.0 + CENTRAL_PROTOCOL_OVERHEAD)

        # Calculate energy
        total_runtime = gov_time + training_time
        energy_data = calculate_energy_cost(total_runtime)

    else:  # Federated training
        # For Random-K, pass the K value to federated training
        k_value = len(tadp_accepted) if approach == 'random-k' else None
        trained_model, comm_cost = federated_training(model, x_test, y_test, accepted, client_data,
                                                    name, approach, mode, job_idx, client_dataset_sizes, k_value)
        training_time = time.time() - start_train

        # Evaluate
        y_proba = trained_model.predict(x_test, verbose=0)
        metrics, cm = get_detailed_metrics(y_test.values, y_proba)

        # Calculate energy
        total_runtime = gov_time + training_time
        energy_data = calculate_energy_cost(total_runtime, len(accepted))

    results = {
        'scenario': name, 'approach': approach, 'mode': mode, 'template': template,
        'num_accepted': len(accepted), 'total_clients': len(client_ids),
        'accuracy': metrics['accuracy'], 'f1': metrics['f1'], 'precision': metrics['precision'],
        'recall': metrics['recall'], 'roc_auc': metrics['roc_auc'],
        'gov_time': gov_time, 'training_time': training_time, 'total_time': total_runtime,
        'comm_cost_mb': comm_cost, 'energy_wh': energy_data['energy_wh'],
        'cost_usd': energy_data['cost_usd'], 'co2_kg': energy_data['co2_kg'],
        'decisions': decisions, 'cutoffs': cutoffs, 'accepted_clients': accepted,
        'confusion_matrix': cm
    }

    print(f"  âœ… Completed: {results['accuracy']:.3f} accuracy, {results['total_time']:.1f}s, {results['comm_cost_mb']:.2f} MB comm")
    return results

# Define all 11 scenarios with proper framework descriptions
scenarios = [
    # Non-governance baselines (5 scenarios)
    {'name': 'NaÃ¯ve Centralized', 'approach': 'baseline', 'mode': 'centralized', 'template': 'AA', 'framework': 'Keras/TensorFlow'},
    {'name': 'NaÃ¯ve Federated (SimpleAvg)', 'approach': 'simpleavg', 'mode': 'federated', 'template': 'AA', 'framework': 'Custom FL (Equal Weight)'},
    {'name': 'Vanilla FedAvg', 'approach': 'fedavg', 'mode': 'federated', 'template': 'AA', 'framework': 'Custom FL (Sample Weight)'},
    {'name': 'Random-K FedAvg', 'approach': 'random-k', 'mode': 'federated', 'template': 'AA', 'framework': 'Custom FL (Random K)'},
    {'name': 'FedProx', 'approach': 'fedprox', 'mode': 'federated', 'template': 'AA', 'framework': 'Custom FL (Proximal)'},

    # TADP governance variants (6 scenarios)
    {'name': 'TADP-AA Centralized', 'approach': 'TADP', 'mode': 'centralized', 'template': 'AA', 'framework': 'Keras/TensorFlow + Governance'},
    {'name': 'TADP-VR Centralized', 'approach': 'TADP', 'mode': 'centralized', 'template': 'VR', 'framework': 'Keras/TensorFlow + Governance'},
    {'name': 'TADP-SDA Centralized', 'approach': 'TADP', 'mode': 'centralized', 'template': 'SDA', 'framework': 'Keras/TensorFlow + Governance'},
    {'name': 'TADP-AA Federated', 'approach': 'TADP', 'mode': 'federated', 'template': 'AA', 'framework': 'Custom FL + Governance'},
    {'name': 'TADP-VR Federated', 'approach': 'TADP', 'mode': 'federated', 'template': 'VR', 'framework': 'Custom FL + Governance'},
    {'name': 'TADP-SDA Federated', 'approach': 'TADP', 'mode': 'federated', 'template': 'SDA', 'framework': 'Custom FL + Governance'},
]

# =========================
# Enhanced Visualization Functions
# =========================
def plot_storage_growth_analysis():
    """Plot comprehensive storage growth analysis"""
    print("\nğŸ“Š Plotting comprehensive storage growth analysis...")

    if not os.path.exists(STORAGE_TRACKING_PATH):
        print("No storage tracking data available")
        return

    storage_df = pd.read_csv(STORAGE_TRACKING_PATH)

    if storage_df.empty:
        print("No storage data to plot")
        return

    # Convert timestamp to datetime
    storage_df['timestamp'] = pd.to_datetime(storage_df['timestamp'])

    # Plot 1: Cumulative storage growth by scenario
    plt.figure(figsize=(15, 10))

    # Subplot 1: Total storage growth over time
    plt.subplot(2, 2, 1)
    for scenario in storage_df['scenario'].unique():
        scenario_data = storage_df[storage_df['scenario'] == scenario]
        plt.plot(scenario_data['timestamp'], scenario_data['total_size_mb'],
                marker='o', linewidth=2, label=scenario, markersize=4)

    plt.xlabel('Time')
    plt.ylabel('Total Storage (MB)')
    plt.title('Cumulative Storage Growth by Scenario')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True, alpha=0.3)
    plt.xticks(rotation=45)

    # Subplot 2: Storage breakdown by component for TADP scenarios
    plt.subplot(2, 2, 2)
    tadp_data = storage_df[storage_df['scenario'].str.contains('TADP')]
    if not tadp_data.empty:
        components = ['central_size_mb', 'local_size_mb', 'overhead_size_mb']
        component_labels = ['Central Ledger', 'Local Ledgers', 'Overhead']
        colors = ['blue', 'orange', 'green']

        bottom = np.zeros(len(tadp_data))
        for i, (comp, label, color) in enumerate(zip(components, component_labels, colors)):
            plt.bar(range(len(tadp_data)), tadp_data[comp], bottom=bottom,
                   label=label, color=color, alpha=0.7)
            bottom += tadp_data[comp].values

        plt.xlabel('Operation Index')
        plt.ylabel('Storage (MB)')
        plt.title('Storage Breakdown for TADP Scenarios')
        plt.legend()
        plt.grid(True, alpha=0.3, axis='y')

    # Subplot 3: Local storage impact per client
    plt.subplot(2, 2, 3)
    client_storage = storage_df.groupby('client')['local_size_mb'].max().sort_values(ascending=False)
    plt.bar(range(len(client_storage)), client_storage.values, color='orange', alpha=0.7)
    plt.xlabel('Client')
    plt.ylabel('Max Local Storage (MB)')
    plt.title('Local Storage Impact per Client')
    plt.xticks(range(len(client_storage)), client_storage.index, rotation=45)
    plt.grid(True, alpha=0.3, axis='y')

    # Subplot 4: Storage growth rate by scenario
    plt.subplot(2, 2, 4)
    growth_rates = []
    scenarios = []
    for scenario in storage_df['scenario'].unique():
        scenario_data = storage_df[storage_df['scenario'] == scenario]
        if len(scenario_data) > 1:
            initial_storage = scenario_data['total_size_mb'].iloc[0]
            final_storage = scenario_data['total_size_mb'].iloc[-1]
            growth_rate = (final_storage - initial_storage) / initial_storage * 100
            growth_rates.append(growth_rate)
            scenarios.append(scenario)

    plt.bar(scenarios, growth_rates, color='red', alpha=0.7)
    plt.xlabel('Scenario')
    plt.ylabel('Storage Growth Rate (%)')
    plt.title('Storage Growth Rate by Scenario')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3, axis='y')

    plt.tight_layout()
    plt.show()

def plot_io_analysis_comprehensive():
    """Plot comprehensive I/O analysis"""
    print("\nğŸ“Š Plotting comprehensive I/O analysis...")

    if not os.path.exists(STORAGE_TRACKING_PATH):
        print("No storage tracking data available")
        return

    storage_df = pd.read_csv(STORAGE_TRACKING_PATH)

    if storage_df.empty:
        print("No I/O data to plot")
        return

    # Plot I/O operations analysis
    plt.figure(figsize=(15, 10))

    # Subplot 1: Operations count by type
    plt.subplot(2, 2, 1)
    operation_counts = storage_df['operation'].value_counts()
    plt.bar(operation_counts.index, operation_counts.values, color='skyblue', alpha=0.7)
    plt.xlabel('Operation Type')
    plt.ylabel('Count')
    plt.title('I/O Operations Count by Type')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3, axis='y')

    # Add value labels
    for i, count in enumerate(operation_counts.values):
        plt.text(i, count + 0.1, str(count), ha='center', va='bottom', fontweight='bold')

    # Subplot 2: Operations by scenario
    plt.subplot(2, 2, 2)
    scenario_ops = storage_df.groupby('scenario')['operation'].count().sort_values(ascending=False)
    plt.bar(range(len(scenario_ops)), scenario_ops.values, color='lightgreen', alpha=0.7)
    plt.xlabel('Scenario')
    plt.ylabel('I/O Operations Count')
    plt.title('I/O Operations by Scenario')
    plt.xticks(range(len(scenario_ops)), scenario_ops.index, rotation=45)
    plt.grid(True, alpha=0.3, axis='y')

    # Subplot 3: Client I/O activity
    plt.subplot(2, 2, 3)
    client_ops = storage_df[storage_df['client'] != 'central']['client'].value_counts().head(10)
    plt.bar(range(len(client_ops)), client_ops.values, color='orange', alpha=0.7)
    plt.xlabel('Client')
    plt.ylabel('I/O Operations Count')
    plt.title('Top 10 Clients by I/O Activity')
    plt.xticks(range(len(client_ops)), client_ops.index, rotation=45)
    plt.grid(True, alpha=0.3, axis='y')

    # Subplot 4: Round-based I/O analysis for federated scenarios
    plt.subplot(2, 2, 4)
    fed_data = storage_df[storage_df['scenario'].str.contains('Federated')]
    if not fed_data.empty and 'round' in fed_data.columns:
        round_ops = fed_data.groupby('round')['operation'].count()
        plt.plot(round_ops.index, round_ops.values, 'o-', linewidth=2, markersize=6)
        plt.xlabel('Round Number')
        plt.ylabel('I/O Operations Count')
        plt.title('I/O Operations per Round (Federated)')
        plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

def plot_communication_analysis(all_results):
    """Plot comprehensive communication analysis"""
    print("\nğŸ“Š Plotting communication analysis...")

    if not all_results:
        print("No results to plot")
        return

    df = pd.DataFrame(all_results)

    # Plot communication analysis
    plt.figure(figsize=(15, 10))

    # Subplot 1: Communication cost by scenario
    plt.subplot(2, 2, 1)
    scenarios = df['scenario'] + " (" + df['mode'] + ")"
    comm_costs = df['comm_cost_mb']

    colors = []
    for _, row in df.iterrows():
        if 'TADP' in row['scenario']:
            colors.append('#2E86AB')
        elif 'Fed' in row['scenario'] or 'Random' in row['scenario']:
            colors.append('#A23B72')
        else:
            colors.append('#F18F01')

    bars = plt.bar(range(len(scenarios)), comm_costs, color=colors, alpha=0.8,
                  edgecolor='black', linewidth=0.5)
    plt.xlabel('Scenario')
    plt.ylabel('Communication Cost (MB)')
    plt.title('Cumulative Communication Cost by Scenario')
    plt.xticks(range(len(scenarios)), scenarios, rotation=45, ha='right')
    plt.grid(True, alpha=0.3, axis='y')

    # Add value labels
    for bar, cost in zip(bars, comm_costs):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(comm_costs)*0.01,
                f'{cost:.1f}', ha='center', va='bottom', fontsize=8, fontweight='bold')

    # Subplot 2: Communication vs Accuracy
    plt.subplot(2, 2, 2)
    scatter = plt.scatter(df['comm_cost_mb'], df['accuracy'], c=df['num_accepted'],
                         cmap='viridis', s=100, alpha=0.7)
    plt.xlabel('Communication Cost (MB)')
    plt.ylabel('Accuracy')
    plt.title('Communication Cost vs Accuracy\n(Color = Number of Accepted Clients)')
    plt.colorbar(scatter, label='Number of Accepted Clients')
    plt.grid(True, alpha=0.3)

    # Subplot 3: Communication efficiency (Accuracy per MB)
    plt.subplot(2, 2, 3)
    efficiency = df['accuracy'] / df['comm_cost_mb']
    efficiency_df = pd.DataFrame({'scenario': scenarios, 'efficiency': efficiency})
    efficiency_df = efficiency_df.sort_values('efficiency', ascending=False)

    bars = plt.bar(range(len(efficiency_df)), efficiency_df['efficiency'],
                  color='green', alpha=0.7)
    plt.xlabel('Scenario')
    plt.ylabel('Accuracy per MB')
    plt.title('Communication Efficiency (Accuracy per MB)')
    plt.xticks(range(len(efficiency_df)), efficiency_df['scenario'], rotation=45, ha='right')
    plt.grid(True, alpha=0.3, axis='y')

    # Subplot 4: Cumulative communication by approach type
    plt.subplot(2, 2, 4)
    approach_comm = df.groupby('approach')['comm_cost_mb'].sum()
    plt.pie(approach_comm.values, labels=approach_comm.index, autopct='%1.1f%%',
            startangle=90, colors=['lightblue', 'lightcoral', 'lightgreen', 'gold'])
    plt.title('Total Communication Cost Distribution by Approach')

    plt.tight_layout()
    plt.show()

def plot_sensitivity_heatmap(sensitivity_df):
    """Plot sensitivity analysis as heatmap"""
    print("\nğŸ“Š Plotting sensitivity analysis heatmap...")

    if sensitivity_df.empty:
        print("No sensitivity data to plot")
        return

    # Create heatmap data
    threshold_data = sensitivity_df[sensitivity_df['type'] == 'threshold']
    weight_data = sensitivity_df[sensitivity_df['type'] == 'weights']

    plt.figure(figsize=(15, 6))

    if not threshold_data.empty:
        plt.subplot(1, 2, 1)
        # Threshold sensitivity heatmap
        acceptance_rates = threshold_data['acceptance_rate'].values
        variations = threshold_data['variation'].values

        # Create a simple heatmap
        im = plt.imshow([acceptance_rates], cmap='RdYlGn', aspect='auto')
        plt.colorbar(im, label='Acceptance Rate')
        plt.xticks(range(len(variations)), variations)
        plt.yticks([0], ['Threshold Variation'])
        plt.xlabel('Threshold Variation')
        plt.title('Threshold Sensitivity Heatmap')

        # Add value annotations
        for i, rate in enumerate(acceptance_rates):
            plt.text(i, 0, f'{rate:.1%}', ha='center', va='center', fontweight='bold', fontsize=12)

    if not weight_data.empty:
        plt.subplot(1, 2, 2)
        # Weight sensitivity heatmap
        acceptance_rates = weight_data['acceptance_rate'].values
        variations = weight_data['variation'].values

        # Create a simple heatmap
        im = plt.imshow([acceptance_rates], cmap='RdYlGn', aspect='auto')
        plt.colorbar(im, label='Acceptance Rate')
        plt.xticks(range(len(variations)), [v.replace('jitter_', 'Jitter ') for v in variations])
        plt.yticks([0], ['Weight Jitter'])
        plt.xlabel('Weight Jitter Factor')
        plt.title('Weight Sensitivity Heatmap')

        # Add value annotations
        for i, rate in enumerate(acceptance_rates):
            plt.text(i, 0, f'{rate:.1%}', ha='center', va='center', fontweight='bold', fontsize=12)

    plt.tight_layout()
    plt.show()

def plot_performance_vs_acceptance(all_results):
    """Plot performance metrics vs acceptance rate"""
    print("\nğŸ“Š Plotting performance vs acceptance rate...")

    if not all_results:
        print("No results to plot")
        return

    df = pd.DataFrame(all_results)

    # Calculate acceptance rate
    df['acceptance_rate'] = df['num_accepted'] / df['total_clients']

    # Separate by mode
    centralized_df = df[df['mode'] == 'centralized']
    federated_df = df[df['mode'] == 'federated']

    # Plot for centralized
    if not centralized_df.empty:
        plt.figure(figsize=(15, 5))
        plt.subplot(1, 3, 1)
        for i, row in centralized_df.iterrows():
            color = 'blue' if 'TADP' in row['scenario'] else 'red'
            plt.scatter(row['acceptance_rate'], row['accuracy'],
                       color=color, s=100, alpha=0.7, label=row['scenario'])
            plt.text(row['acceptance_rate'], row['accuracy'], row['scenario'],
                    fontsize=8, alpha=0.7)

        plt.xlabel('Acceptance Rate')
        plt.ylabel('Accuracy')
        plt.title('Accuracy vs Acceptance Rate (Centralized)')
        plt.grid(True, alpha=0.3)

    # Plot for federated
    if not federated_df.empty:
        plt.subplot(1, 3, 2)
        for i, row in federated_df.iterrows():
            color = 'blue' if 'TADP' in row['scenario'] else 'red'
            plt.scatter(row['acceptance_rate'], row['accuracy'],
                       color=color, s=100, alpha=0.7, label=row['scenario'])
            plt.text(row['acceptance_rate'], row['accuracy'], row['scenario'],
                    fontsize=8, alpha=0.7)

        plt.xlabel('Acceptance Rate')
        plt.ylabel('Accuracy')
        plt.title('Accuracy vs Acceptance Rate (Federated)')
        plt.grid(True, alpha=0.3)

    # Plot F1 score vs acceptance rate
    plt.subplot(1, 3, 3)
    for i, row in df.iterrows():
        color = 'blue' if 'TADP' in row['scenario'] else 'red'
        marker = 'o' if row['mode'] == 'centralized' else 's'
        plt.scatter(row['acceptance_rate'], row['f1'],
                   color=color, s=100, alpha=0.7, marker=marker, label=row['scenario'])

    plt.xlabel('Acceptance Rate')
    plt.ylabel('F1 Score')
    plt.title('F1 Score vs Acceptance Rate\n(Circles=Centralized, Squares=Federated)')
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

def plot_energy_fairness_analysis(all_results):
    """Plot energy fairness analysis for TADP scenarios"""
    print("\nğŸ“Š Plotting energy fairness analysis...")

    if not all_results:
        print("No results to plot")
        return

    df = pd.DataFrame(all_results)

    # Filter TADP scenarios
    tadp_df = df[df['approach'] == 'TADP']

    if tadp_df.empty:
        print("No TADP scenarios for energy analysis")
        return

    # Calculate average energy for TADP scenarios
    tadp_federated_avg = tadp_df[tadp_df['mode'] == 'federated']['energy_wh'].mean()
    tadp_centralized_avg = tadp_df[tadp_df['mode'] == 'centralized']['energy_wh'].mean()

    # Plot energy comparison
    plt.figure(figsize=(12, 6))

    # Subplot 1: Average energy comparison
    plt.subplot(1, 2, 1)
    approaches = ['TADP Federated', 'TADP Centralized']
    energies = [tadp_federated_avg, tadp_centralized_avg]
    colors = ['lightblue', 'lightcoral']

    bars = plt.bar(approaches, energies, color=colors, alpha=0.7, edgecolor='black')
    plt.ylabel('Average Energy Consumption (Wh)')
    plt.title('Average Energy Consumption:\nTADP Federated vs Centralized')
    plt.grid(True, alpha=0.3, axis='y')

    # Add value labels
    for bar, energy in zip(bars, energies):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(energies)*0.01,
                f'{energy:.1f} Wh', ha='center', va='bottom', fontweight='bold')

    # Subplot 2: Energy vs Accuracy for all TADP scenarios
    plt.subplot(1, 2, 2)
    for i, row in tadp_df.iterrows():
        color = 'blue' if row['mode'] == 'federated' else 'red'
        marker = 'o' if row['mode'] == 'federated' else 's'
        plt.scatter(row['energy_wh'], row['accuracy'],
                   color=color, s=100, alpha=0.7, marker=marker, label=row['scenario'])
        plt.text(row['energy_wh'], row['accuracy'], row['scenario'],
                fontsize=8, alpha=0.7)

    plt.xlabel('Energy Consumption (Wh)')
    plt.ylabel('Accuracy')
    plt.title('Energy vs Accuracy for TADP Scenarios\n(Blue=Federated, Red=Centralized)')
    plt.grid(True, alpha=0.3)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout()
    plt.show()

    print(f"\nğŸ“ˆ Energy Fairness Summary:")
    print(f"   â€¢ TADP Federated Average: {tadp_federated_avg:.2f} Wh")
    print(f"   â€¢ TADP Centralized Average: {tadp_centralized_avg:.2f} Wh")
    print(f"   â€¢ Energy Ratio (Federated/Centralized): {tadp_federated_avg/tadp_centralized_avg:.2f}")

# =========================
# Analysis Functions
# =========================
def performance_per_client_analysis(all_results, client_data):
    """Analyze performance metrics per client across scenarios"""
    print("\n" + "="*80)
    print("ğŸ“Š PERFORMANCE PER CLIENT ANALYSIS")
    print("="*80)

    client_performance = []

    for result in all_results:
        scenario = result['scenario']
        approach = result['approach']
        decisions = result['decisions']

        for client_id, decision_info in decisions.items():
            initial_action, final_action, pscore, dim_vals, profile = decision_info
            dim1, dim2, dim3, dim4, dim5, dim6 = dim_vals

            client_performance.append({
                'scenario': scenario,
                'approach': approach,
                'client': client_id,
                'pscore': pscore,
                'dim1': dim1,
                'dim2': dim2,
                'dim3': dim3,
                'dim4': dim4,
                'dim5': dim5,
                'dim6': dim6,
                'profile': profile,
                'initial_action': initial_action,
                'final_action': final_action,
                'accepted': final_action == 'ACCEPT',
                'dataset_size': len(client_data[client_id])
            })

    perf_df = pd.DataFrame(client_performance)

    # Display client performance summary
    print("\nClient Performance Summary:")
    summary = perf_df.groupby(['scenario', 'approach']).agg({
        'client': 'count',
        'pscore': 'mean',
        'accepted': 'sum',
        'dataset_size': 'mean'
    }).round(3)

    summary['acceptance_rate'] = (summary['accepted'] / summary['client'] * 100).round(1)
    summary.columns = ['Total_Clients', 'Avg_Pscore', 'Accepted', 'Avg_Dataset_Size', 'Acceptance_Rate%']

    try:
        display(summary.style.background_gradient(subset=['Avg_Pscore', 'Acceptance_Rate%'], cmap='RdYlGn'))
    except:
        print(summary)

    # Plot client P-score distribution
    plt.figure(figsize=(12, 6))
    for approach in perf_df['approach'].unique():
        subset = perf_df[perf_df['approach'] == approach]
        plt.hist(subset['pscore'], alpha=0.6, label=approach, bins=20)

    plt.axvline(x=HIGH_CUT, color='r', linestyle='--', label=f'High Cutoff ({HIGH_CUT})')
    plt.axvline(x=GOOD_CUT, color='orange', linestyle='--', label=f'Good Cutoff ({GOOD_CUT})')
    plt.xlabel('P-score')
    plt.ylabel('Number of Clients')
    plt.title('Distribution of Client P-scores by Approach')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    return perf_df

def run_ablation_studies(client_ids, client_data, template='VR'):
    """Run ablation studies by zeroing out each dimension"""
    print("\n" + "="*80)
    print("ğŸ”¥ ABLATION ANALYSIS - DIMENSION IMPACT")
    print("="*80)

    ablation_results = []
    original_weights = WEIGHTS_PSCORE.copy()

    # Baseline (all dimensions active)
    print("  Running baseline (all dimensions active)...")
    baseline_accepted, baseline_decisions, baseline_cutoffs = run_tadp_governance(
        client_ids, client_data, "Ablation_Baseline", "TADP", "federated", 0, template
    )
    baseline_acceptance = len(baseline_accepted) / len(client_ids)

    ablation_results.append({
        'ablated_dimension': 'NONE (Baseline)',
        'acceptance_rate': baseline_acceptance,
        'num_accepted': len(baseline_accepted),
        'avg_pscore': np.mean([d[2] for d in baseline_decisions.values()]),
        'impact_vs_baseline': 0.0
    })

    # Ablate each dimension
    dimensions = list(WEIGHTS_PSCORE.keys())
    dimension_names = {
        'dim1': 'Data Source Reliability',
        'dim2': 'Clinical Completeness',
        'dim3': 'Documentation Practices',
        'dim4': 'Time Freshness',
        'dim5': 'Regulatory Compliance',
        'dim6': 'Usage Constraints'
    }

    for dim in dimensions:
        print(f"  Ablating {dimension_names[dim]}...")

        # Create modified weights (zero out this dimension)
        modified_weights = original_weights.copy()
        modified_weights[dim] = 0.0

        # Normalize remaining weights
        total_weight = sum(modified_weights.values())
        for k in modified_weights:
            modified_weights[k] /= total_weight

        # Calculate modified P-scores
        modified_pscores = {}
        for cid in client_ids:
            pscore, dim_vals, profile = calculate_pscore(cid, client_data[cid])
            # Apply modified weights to dimension values
            modified_pscore = sum(d * modified_weights[f'dim{i+1}'] for i, d in enumerate(dim_vals))
            modified_pscores[cid] = modified_pscore

        # Apply governance with modified P-scores
        high_cut, good_cut = dynamic_cutoffs(list(modified_pscores.values()), template)
        accepted = []
        for cid, pscore in modified_pscores.items():
            action = decide_action(pscore, high_cut, good_cut, template)
            if action == 'ACCEPT' or (action == 'REVIEW' and np.random.random() < 0.7):
                accepted.append(cid)

        acceptance_rate = len(accepted) / len(client_ids)
        impact = (acceptance_rate - baseline_acceptance) / baseline_acceptance * 100

        ablation_results.append({
            'ablated_dimension': dimension_names[dim],
            'acceptance_rate': acceptance_rate,
            'num_accepted': len(accepted),
            'avg_pscore': np.mean(list(modified_pscores.values())),
            'impact_vs_baseline': impact
        })

    ablation_df = pd.DataFrame(ablation_results)

    # Display ablation results
    print("\nAblation Study Results:")
    try:
        display(ablation_df.style.background_gradient(
            subset=['acceptance_rate', 'impact_vs_baseline'],
            cmap='RdYlGn_r'
        ).format({
            'acceptance_rate': '{:.1%}',
            'avg_pscore': '{:.2f}',
            'impact_vs_baseline': '{:.1f}%'
        }))
    except:
        print(ablation_df.round(3))

    # Plot ablation impact
    plt.figure(figsize=(10, 6))
    colors = ['green' if x >= 0 else 'red' for x in ablation_df['impact_vs_baseline']]
    bars = plt.bar(ablation_df['ablated_dimension'], ablation_df['impact_vs_baseline'], color=colors, alpha=0.7)
    plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    plt.ylabel('Impact vs Baseline (%)')
    plt.title('Ablation Study: Impact of Zeroing Out Each Dimension\n(Positive = More Acceptances, Negative = Fewer Acceptances)')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, alpha=0.3, axis='y')

    # Add value labels
    for bar, impact in zip(bars, ablation_df['impact_vs_baseline']):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + (1 if bar.get_height() >= 0 else -2),
                f'{impact:+.1f}%', ha='center', va='bottom' if bar.get_height() >= 0 else 'top', fontweight='bold')

    plt.tight_layout()
    plt.show()

    return ablation_df

def run_sensitivity_analysis(client_ids, client_data, template='VR'):
    """Run sensitivity analysis with threshold and weight variations"""
    print("\n" + "="*80)
    print("ğŸ“Š SENSITIVITY ANALYSIS")
    print("="*80)

    sensitivity_results = []

    # Threshold variations
    print("  Testing threshold variations...")
    threshold_variations = [-0.5, -0.25, 0.0, 0.25, 0.5]

    for threshold_shift in threshold_variations:
        high_cut_modified = HIGH_CUT + threshold_shift
        good_cut_modified = GOOD_CUT + threshold_shift

        # Calculate baseline P-scores
        pscores = [calculate_pscore(cid, client_data[cid])[0] for cid in client_ids]

        # Apply modified thresholds
        accepted = []
        for cid, pscore in zip(client_ids, pscores):
            action = decide_action(pscore, high_cut_modified, good_cut_modified, template)
            if action == 'ACCEPT' or (action == 'REVIEW' and np.random.random() < 0.7):
                accepted.append(cid)

        sensitivity_results.append({
            'type': 'threshold',
            'variation': f'{threshold_shift:+.2f}',
            'high_cut': high_cut_modified,
            'good_cut': good_cut_modified,
            'acceptance_rate': len(accepted) / len(client_ids),
            'num_accepted': len(accepted),
            'avg_pscore': np.mean(pscores)
        })

    # Weight variations (jitter)
    print("  Testing weight variations...")
    jitter_factors = [0.0, 0.1, 0.2, 0.3]

    for jitter_factor in jitter_factors:
        jittered_weights = WEIGHTS_PSCORE.copy()

        # Apply jitter to weights
        for dim in jittered_weights:
            jitter = np.random.normal(0, jitter_factor * jittered_weights[dim])
            jittered_weights[dim] = max(0.01, jittered_weights[dim] + jitter)

        # Normalize weights
        total_weight = sum(jittered_weights.values())
        for dim in jittered_weights:
            jittered_weights[dim] /= total_weight

        # Calculate modified P-scores
        modified_pscores = []
        for cid in client_ids:
            pscore, dim_vals, _ = calculate_pscore(cid, client_data[cid])
            # Apply jittered weights
            modified_pscore = sum(d * jittered_weights[f'dim{i+1}'] for i, d in enumerate(dim_vals))
            modified_pscores.append(modified_pscore)

        # Apply governance
        high_cut, good_cut = dynamic_cutoffs(modified_pscores, template)
        accepted = []
        for cid, pscore in zip(client_ids, modified_pscores):
            action = decide_action(pscore, high_cut, good_cut, template)
            if action == 'ACCEPT' or (action == 'REVIEW' and np.random.random() < 0.7):
                accepted.append(cid)

        sensitivity_results.append({
            'type': 'weights',
            'variation': f'jitter_{jitter_factor}',
            'high_cut': high_cut,
            'good_cut': good_cut,
            'acceptance_rate': len(accepted) / len(client_ids),
            'num_accepted': len(accepted),
            'avg_pscore': np.mean(modified_pscores)
        })

    sensitivity_df = pd.DataFrame(sensitivity_results)

    # Display sensitivity results
    print("\nSensitivity Analysis Results:")
    try:
        display(sensitivity_df.style.background_gradient(
            subset=['acceptance_rate'],
            cmap='RdYlGn'
        ).format({
            'acceptance_rate': '{:.1%}',
            'avg_pscore': '{:.2f}',
            'high_cut': '{:.2f}',
            'good_cut': '{:.2f}'
        }))
    except:
        print(sensitivity_df.round(3))

    # Plot sensitivity results using heatmap
    plot_sensitivity_heatmap(sensitivity_df)

    return sensitivity_df

def create_tadp_governance_summary(all_results):
    """Create comprehensive TADP governance summary table"""
    print("\n" + "="*80)
    print("ğŸ›ï¸  TADP GOVERNANCE SUMMARY")
    print("="*80)

    tadp_results = [r for r in all_results if 'TADP' in r['approach']]

    if not tadp_results:
        print("No TADP results available for governance summary")
        return

    governance_summary = []

    for result in tadp_results:
        decisions = result['decisions']
        total_clients = len(decisions)

        # Count decisions
        accept_count = sum(1 for d in decisions.values() if d[1] == 'ACCEPT')
        review_count = sum(1 for d in decisions.values() if d[1] == 'REVIEW')
        quarantine_count = sum(1 for d in decisions.values() if d[1] == 'QUARANTINE')

        # Calculate P-score statistics
        pscores = [d[2] for d in decisions.values()]
        accepted_pscores = [d[2] for d in decisions.values() if d[1] == 'ACCEPT']

        governance_summary.append({
            'Scenario': result['scenario'],
            'Template': result['template'],
            'Total_Clients': total_clients,
            'Accepted': accept_count,
            'Under_Review': review_count,
            'Quarantined': quarantine_count,
            'Acceptance_Rate': accept_count / total_clients,
            'Avg_Pscore_All': np.mean(pscores),
            'Avg_Pscore_Accepted': np.mean(accepted_pscores) if accepted_pscores else 0,
            'Model_Accuracy': result['accuracy'],
            'Gov_Time_s': result['gov_time']
        })

    gov_df = pd.DataFrame(governance_summary)

    # Display governance summary
    try:
        styled_gov = gov_df.style.background_gradient(
            subset=['Acceptance_Rate', 'Avg_Pscore_All', 'Avg_Pscore_Accepted', 'Model_Accuracy'],
            cmap='RdYlGn'
        ).background_gradient(
            subset=['Gov_Time_s'],
            cmap='YlOrBr'
        ).format({
            'Acceptance_Rate': '{:.1%}',
            'Avg_Pscore_All': '{:.2f}',
            'Avg_Pscore_Accepted': '{:.2f}',
            'Model_Accuracy': '{:.2%}',
            'Gov_Time_s': '{:.2f}'
        })

        styled_gov = styled_gov.set_caption("TADP Governance Performance Summary")
        display(styled_gov)
    except:
        print(gov_df.round(3))

    # Plot governance decision distribution
    plt.figure(figsize=(12, 6))

    scenarios = gov_df['Scenario']
    acceptance_rates = gov_df['Acceptance_Rate'] * 100

    plt.bar(scenarios, acceptance_rates, color='green', alpha=0.7, label='Accepted')

    # Add model accuracy as line plot
    ax2 = plt.twinx()
    ax2.plot(scenarios, gov_df['Model_Accuracy'] * 100, 'ro-', linewidth=2, markersize=6, label='Model Accuracy')
    ax2.set_ylabel('Model Accuracy (%)', color='red')
    ax2.tick_params(axis='y', labelcolor='red')

    plt.xlabel('Scenario')
    plt.ylabel('Acceptance Rate (%)')
    plt.title('TADP Governance: Acceptance Rate vs Model Accuracy')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, alpha=0.3, axis='y')

    # Combine legends
    lines1, labels1 = plt.gca().get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    plt.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

    plt.tight_layout()
    plt.show()

    return gov_df

# =========================
# Basic Visualization Functions (for compatibility)
# =========================
def styled_performance_table(df):
    """Create styled performance table with color gradients"""
    perf_cols = ["accuracy", "precision", "recall", "f1", "roc_auc", "total_time"]
    view_cols = ["scenario", "approach"] + perf_cols

    tbl = df[view_cols].copy()
    tbl.columns = ["Scenario", "Approach", "Accuracy", "Precision", "Recall", "F1", "ROC AUC", "Time (s)"]

    fmt = {
        "Accuracy": "{:.2%}",
        "Precision": "{:.2%}",
        "Recall": "{:.2%}",
        "F1": "{:.2%}",
        "ROC AUC": "{:.2%}",
        "Time (s)": "{:.1f}"
    }

    try:
        cm = tbl.style.background_gradient(
            subset=["Accuracy", "Precision", "Recall", "F1", "ROC AUC"],
            cmap="Greens"
        ).background_gradient(
            subset=["Time (s)"],
            cmap="YlOrBr"
        ).format(fmt)

        cm = cm.set_caption("Table 1: Model Performance Metrics by Scenario (All 11 Scenarios)")
        return cm
    except Exception:
        print(tbl.to_string(index=False))
        return tbl

def styled_resource_table(df):
    """Create styled resource usage table with color gradients"""
    res_cols = ["comm_cost_mb", "energy_wh", "cost_usd", "co2_kg"]
    view_cols = ["scenario", "approach"] + res_cols

    tbl = df[view_cols].copy()
    tbl.columns = ["Scenario", "Approach", "Comm (MB)", "Energy (Wh)", "Cost (USD)", "CO2 (kg)"]

    fmt = {
        "Comm (MB)": "{:.1f}",
        "Energy (Wh)": "{:.1f}",
        "Cost (USD)": "${:.3f}",
        "CO2 (kg)": "{:.3f}"
    }

    try:
        cm = tbl.style.background_gradient(
            subset=["Comm (MB)", "Energy (Wh)", "CO2 (kg)"],
            cmap="Reds"
        ).background_gradient(
            subset=["Cost (USD)"],
            cmap="Oranges"
        ).format(fmt)

        cm = cm.set_caption("Table 2: Resource Usage and Costs by Scenario (All 11 Scenarios)")
        return cm
    except Exception:
        print(tbl.to_string(index=False))
        return tbl

def plot_performance_comparison(all_results):
    """Plot accuracy and F1 scores across all scenarios"""
    if not all_results:
        print("No results to plot")
        return

    df = pd.DataFrame(all_results)
    scenarios = df['scenario'] + " (" + df['mode'] + ")"
    accuracies = df['accuracy']
    f1_scores = df['f1']

    # Create color coding
    colors = []
    for _, row in df.iterrows():
        if 'TADP' in row['scenario']:
            colors.append('#2E86AB')  # Blue for TADP
        elif 'Fed' in row['scenario'] or 'Random' in row['scenario']:
            colors.append('#A23B72')  # Purple for FL methods
        else:
            colors.append('#F18F01')  # Orange for baselines

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # Accuracy plot
    bars1 = ax1.bar(range(len(scenarios)), accuracies, color=colors, alpha=0.8,
                   edgecolor='black', linewidth=0.5)
    ax1.set_title('Model Accuracy by Scenario', fontsize=14, fontweight='bold', pad=20)
    ax1.set_ylabel('Accuracy', fontweight='bold')
    ax1.set_xticks(range(len(scenarios)))
    ax1.set_xticklabels(scenarios, rotation=45, ha='right', fontsize=9)
    ax1.grid(True, alpha=0.3, axis='y')
    ax1.set_ylim(0, 1.0)

    # Add value labels
    for i, (bar, acc) in enumerate(zip(bars1, accuracies)):
        ax1.text(bar.get_x() + bar.get_width()/2., acc + 0.01,
                f'{acc:.3f}', ha='center', va='bottom', fontsize=8, fontweight='bold')

    # F1 score plot
    bars2 = ax2.bar(range(len(scenarios)), f1_scores, color=colors, alpha=0.8,
                   edgecolor='black', linewidth=0.5)
    ax2.set_title('F1 Score by Scenario', fontsize=14, fontweight='bold', pad=20)
    ax2.set_ylabel('F1 Score', fontweight='bold')
    ax2.set_xticks(range(len(scenarios)))
    ax2.set_xticklabels(scenarios, rotation=45, ha='right', fontsize=9)
    ax2.grid(True, alpha=0.3, axis='y')
    ax2.set_ylim(0, 1.0)

    # Add value labels
    for i, (bar, f1) in enumerate(zip(bars2, f1_scores)):
        ax2.text(bar.get_x() + bar.get_width()/2., f1 + 0.01,
                f'{f1:.3f}', ha='center', va='bottom', fontsize=8, fontweight='bold')

    # Add legend
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(facecolor='#F18F01', label='Baselines'),
        Patch(facecolor='#A23B72', label='FL Methods'),
        Patch(facecolor='#2E86AB', label='TADP Methods')
    ]
    ax2.legend(handles=legend_elements, loc='upper right')

    plt.tight_layout()
    plt.savefig('performance_comparison.png', dpi=300, bbox_inches='tight', facecolor='white')
    plt.show()

def plot_energy_communication_costs(all_results):
    """Plot energy and communication costs"""
    if not all_results:
        print("No results to plot")
        return

    df = pd.DataFrame(all_results)
    scenarios = df['scenario'] + " (" + df['mode'] + ")"
    energy_costs = df['energy_wh']
    comm_costs = df['comm_cost_mb']

    # Create color coding
    colors = []
    for _, row in df.iterrows():
        if 'TADP' in row['scenario']:
            colors.append('#2E86AB')
        elif 'Fed' in row['scenario'] or 'Random' in row['scenario']:
            colors.append('#A23B72')
        else:
            colors.append('#F18F01')

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # Energy costs
    bars1 = ax1.bar(range(len(scenarios)), energy_costs, color=colors, alpha=0.8,
                   edgecolor='black', linewidth=0.5)
    ax1.set_title('Energy Consumption by Scenario', fontsize=14, fontweight='bold', pad=20)
    ax1.set_ylabel('Energy (Wh)', fontweight='bold')
    ax1.set_xticks(range(len(scenarios)))
    ax1.set_xticklabels(scenarios, rotation=45, ha='right', fontsize=9)
    ax1.grid(True, alpha=0.3, axis='y')

    # Add value labels
    for i, (bar, energy) in enumerate(zip(bars1, energy_costs)):
        ax1.text(bar.get_x() + bar.get_width()/2., energy + max(energy_costs)*0.01,
                f'{energy:.1f}', ha='center', va='bottom', fontsize=8, fontweight='bold')

    # Communication costs
    bars2 = ax2.bar(range(len(scenarios)), comm_costs, color=colors, alpha=0.8,
                   edgecolor='black', linewidth=0.5)
    ax2.set_title('Communication Costs by Scenario', fontsize=14, fontweight='bold', pad=20)
    ax2.set_ylabel('Communication (MB)', fontweight='bold')
    ax2.set_xticks(range(len(scenarios)))
    ax2.set_xticklabels(scenarios, rotation=45, ha='right', fontsize=9)
    ax2.grid(True, alpha=0.3, axis='y')

    # Add value labels
    for i, (bar, comm) in enumerate(zip(bars2, comm_costs)):
        ax2.text(bar.get_x() + bar.get_width()/2., comm + max(comm_costs)*0.01,
                f'{comm:.1f}', ha='center', va='bottom', fontsize=8, fontweight='bold')

    plt.tight_layout()
    plt.savefig('energy_communication_costs.png', dpi=300, bbox_inches='tight', facecolor='white')
    plt.show()

# =========================
# Main Execution
# =========================
def main():
    """Main execution function"""
    print("ğŸš€ Starting TADP: Trustworthy AI Data Preparation Governance & Evaluation Pipeline")
    print("=" * 80)
    print("COMPREHENSIVE ANALYSIS SUITE ACTIVATED:")
    print("â€¢ Performance per client analysis")
    print("â€¢ Ablation studies (dimension impact)")
    print("â€¢ Sensitivity analysis (thresholds & weights)")
    print("â€¢ TADP governance summary tables")
    print("â€¢ Enhanced storage and I/O analysis")
    print("â€¢ Communication cost analysis")
    print("â€¢ Performance vs acceptance rate analysis")
    print("â€¢ Energy fairness analysis")
    print("=" * 80)

    # Reset ledgers for clean start
    reset_ledgers()

    # Load and preprocess data
    print("\nğŸ“Š Loading and preprocessing diabetes dataset...")
    df = load_diabetes_csv()
    df = preprocess_diabetes(df)
    print(f"   Dataset shape: {df.shape}")
    print(f"   Target distribution:\n{df['readmitted'].value_counts()}")

    # Split into clients
    print(f"\nğŸ”€ Splitting data into {NUM_CLIENTS} non-IID clients...")
    client_ids, client_data = split_clients(df, NUM_CLIENTS)
    print(f"   Client sizes: {[len(client_data[cid]) for cid in client_ids]}")

    # Estimate dataset sizes for communication calculations
    print("\nğŸ“ Estimating dataset sizes for communication costs...")
    client_dataset_sizes = estimate_dataset_size_mb(client_data)
    for cid, size in client_dataset_sizes.items():
        print(f"   Client {cid}: {size:.2f} MB")

    # Preprocess for ML
    print("\nğŸ”§ Preprocessing data for machine learning...")
    global preprocessor
    X_train, X_test, y_train, y_test, preprocessor, cv = preprocess_data(client_data)
    print(f"   Training set: {X_train.shape}, Test set: {X_test.shape}")

    print("\nğŸ¯ Running all 11 scenarios with comprehensive analysis...")

    # Run scenarios
    all_results = []

    for i, scenario in enumerate(scenarios):
        try:
            print(f"\n{'='*60}")
            print(f"Scenario {i+1}/{len(scenarios)}: {scenario['name']}")
            print(f"{'='*60}")

            result = run_scenario(
                scenario['name'], scenario['approach'], scenario['mode'], scenario['template'],
                client_ids, client_data, client_dataset_sizes, X_test, y_test, i+1
            )
            all_results.append(result)
            print(f"âœ… {scenario['name']} completed successfully")
        except Exception as e:
            print(f"âŒ Scenario {scenario['name']} failed: {e}")
            import traceback
            traceback.print_exc()
            continue

    print(f"\nâœ… Completed {len(all_results)}/{len(scenarios)} scenarios")

    # =========================
    # COMPREHENSIVE ANALYSIS SUITE
    # =========================

    if all_results:
        print("\n" + "="*100)
        print("ğŸ“Š COMPREHENSIVE ANALYSIS SUITE")
        print("=" * 100)

        # Convert to DataFrame for analysis
        results_df = pd.DataFrame(all_results)

        # 1. Basic performance and resource tables
        print("\n1. BASIC PERFORMANCE AND RESOURCE TABLES")
        try:
            display(styled_performance_table(results_df))
            display(styled_resource_table(results_df))
        except:
            print("Performance Table:")
            print(results_df[['scenario', 'approach', 'mode', 'accuracy', 'f1', 'precision', 'recall', 'roc_auc', 'total_time']].to_string(index=False))
            print("\nResource Table:")
            print(results_df[['scenario', 'approach', 'mode', 'comm_cost_mb', 'energy_wh', 'cost_usd', 'co2_kg']].to_string(index=False))

        # 2. Enhanced storage and I/O analysis
        print("\n2. ENHANCED STORAGE AND I/O ANALYSIS")
        plot_storage_growth_analysis()
        plot_io_analysis_comprehensive()

        # 3. Communication analysis
        print("\n3. COMMUNICATION ANALYSIS")
        plot_communication_analysis(all_results)

        # 4. Performance vs acceptance rate
        print("\n4. PERFORMANCE VS ACCEPTANCE RATE ANALYSIS")
        plot_performance_vs_acceptance(all_results)

        # 5. Energy fairness analysis
        print("\n5. ENERGY FAIRNESS ANALYSIS")
        plot_energy_fairness_analysis(all_results)

        # 6. Performance per client analysis
        print("\n6. PERFORMANCE PER CLIENT ANALYSIS")
        client_perf_df = performance_per_client_analysis(all_results, client_data)

        # 7. Ablation studies
        print("\n7. ABLATION ANALYSIS")
        ablation_df = run_ablation_studies(client_ids, client_data)

        # 8. Sensitivity analysis
        print("\n8. SENSITIVITY ANALYSIS")
        sensitivity_df = run_sensitivity_analysis(client_ids, client_data)

        # 9. TADP governance summary
        print("\n9. TADP GOVERNANCE SUMMARY")
        gov_summary_df = create_tadp_governance_summary(all_results)

        # 10. Basic visualizations (for compatibility)
        print("\n10. BASIC VISUALIZATIONS")
        plot_performance_comparison(all_results)
        plot_energy_communication_costs(all_results)

        # Final comprehensive summary
        print("\n" + "="*100)
        print("ğŸ¯ KEY INSIGHTS SUMMARY")
        print("=" * 100)

        # Calculate key metrics
        tadp_scenarios = [r for r in all_results if 'TADP' in r['approach']]
        baseline_scenarios = [r for r in all_results if 'baseline' in r['approach'] or 'NaÃ¯ve' in r['scenario']]

        if tadp_scenarios and baseline_scenarios:
            avg_tadp_accuracy = np.mean([r['accuracy'] for r in tadp_scenarios])
            avg_baseline_accuracy = np.mean([r['accuracy'] for r in baseline_scenarios])
            accuracy_improvement = ((avg_tadp_accuracy - avg_baseline_accuracy) / avg_baseline_accuracy) * 100

            avg_tadp_comm = np.mean([r['comm_cost_mb'] for r in tadp_scenarios])
            avg_baseline_comm = np.mean([r['comm_cost_mb'] for r in baseline_scenarios])
            comm_reduction = ((avg_baseline_comm - avg_tadp_comm) / avg_baseline_comm) * 100

            print(f"ğŸ“ˆ Performance Insights:")
            print(f"   â€¢ Average TADP Accuracy: {avg_tadp_accuracy:.1%}")
            print(f"   â€¢ Accuracy Improvement: {accuracy_improvement:+.1f}%")
            print(f"   â€¢ Communication Reduction: {comm_reduction:+.1f}%")

        # Most critical dimension from ablation
        if not ablation_df.empty:
            most_critical = ablation_df.loc[ablation_df['impact_vs_baseline'].idxmin()]
            print(f"\nğŸ” Governance Insights:")
            print(f"   â€¢ Most Critical Dimension: {most_critical['ablated_dimension']}")
            print(f"   â€¢ Impact when removed: {most_critical['impact_vs_baseline']:.1f}%")

        # Best performing TADP template
        if not gov_summary_df.empty:
            best_template = gov_summary_df.loc[gov_summary_df['Model_Accuracy'].idxmax()]
            print(f"\nğŸ† Best Performing TADP Template:")
            print(f"   â€¢ Scenario: {best_template['Scenario']}")
            print(f"   â€¢ Accuracy: {best_template['Model_Accuracy']:.1%}")
            print(f"   â€¢ Acceptance Rate: {best_template['Acceptance_Rate']:.1%}")

        # Final storage accounting
        print("\nğŸ’¾ FINAL STORAGE ACCOUNTING")
        print("="*50)
        central_size = _sizeof_mb(CENTRAL_LEDGER_PATH)
        local_size = sum(_sizeof_mb(f"./ledgers/local/{cid}_ledger.csv") for cid in client_ids)
        overhead_size = _sizeof_mb(OVERHEAD_PATH)
        total_size = central_size + local_size + overhead_size

        print(f"Central Ledger: {central_size:.6f} MB")
        print(f"Local Ledgers: {local_size:.6f} MB")
        print(f"Overhead: {overhead_size:.6f} MB")
        print(f"TOTAL: {total_size:.6f} MB")

    print("\nğŸ‰ TADP COMPREHENSIVE ANALYSIS COMPLETED SUCCESSFULLY!")

if __name__ == "__main__":
    main()
```
